```python
data = [x for x in np.arange(0, 1.0025, 0.0025)] 

df = pd.DataFrame(data, columns=['talent'])
df
```

```python
df.insert(loc=len(df.columns), column='exam', value= (30 + 70*df['talent']))
df
```

```python
df.insert(loc=len(df.columns), column='education', value= 0)
df
```

```python
for i in range(len(df)):
    if (df['exam'][i] > 80):
        df['education'][i] = 1
```

```python
df
```

```python
df['education'].value_counts()
```

```python
df.insert(loc=len(df.columns), column='salary', value = (30000 + 20000 * df['talent'] + 25000*df['education']))
df
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
ser1 = df[df['education'] == 1].salary.tolist()
statistics.mean(ser1)
```

```python
ser0 = df[df['education'] == 0].salary.tolist()
statistics.mean(ser0)
```

```python
statistics.mean(ser1) - statistics.mean(ser0)
```

```python
df1 = df[(df['exam'] >= 75) & (df['exam'] <= 85)]
df1
```

```python
ser1_new = df1[df1['education'] == 1].salary.tolist()
statistics.mean(ser1_new)
```

```python
ser0_new = df1[df1['education'] == 0].salary.tolist()
statistics.mean(ser0_new)
```

```python
statistics.mean(ser1_new) - statistics.mean(ser0_new)
```
